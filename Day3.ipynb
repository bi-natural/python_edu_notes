```python
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.3)

results = []
for c in [0.001, 0.01, 0.1, 1., 10., 100., 1000.]:
    model = LogisticRegression(C=c)
    model.fit(X_tr, y_tr)
    score_tr = model.score(X_tr, y_tr)
    score_te = model.score(X_te, y_te)
    results.append((c, score_tr, score_te))
```

```python
report = DataFrame(results)
report.columns = ['C', 'train', 'test']
report = report.set_index('C')
plt.rc('font', family='Monaco')
report.plot(logx=True, style=['g*--', 'ro--'])
```

# 연습
스케일의 효과를 확인

```python
X_tr, X_te, Xstd_tr, Xstd_te, y_tr, y_te = train_test_split(X, Xstd, y, test_size=0.3)
```

```python
model = LogisticRegression(C=1.0)
```

```python
model.fit(X_tr, y_tr)
```

```python
model.score(X_te, y_te), model.score(Xstd_te, y_te)
```

```python
X_tr, X_te, Xstd_tr, Xstd_te, y_tr, y_te = train_test_split(X, Xstd, y, test_size=0.3)
results = { 'c': [], 'normal': [], 'std': []}
for c in [0.01, 0.1, 1., 10., 100., 1000.]:
    results['c'].append(c)
    
    model = LogisticRegression(C=c)
    model.fit(X_tr, y_tr)
    results['normal'].append((model.score(X_tr, y_tr), model.score(X_te, y_te)))
    
    model.fit(Xstd_tr, y_tr)
    results['std'].append((model.score(Xstd_tr, y_tr), model.score(Xstd_te, y_te)))        
    
```

```python
DataFrame(results)
```

```python
rep = DataFrame(results)
rep = rep.set_index('c')
rep.plot()
```
