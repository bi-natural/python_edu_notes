```python
def make_logic_gate(w, b):
    def logic_gate(x1, x2):
        x = np.array([x1, x2])        
        z = np.dot(x, w)
        y = 1 if z > b else -1
        return 1 if y > 0 else 0
    return logic_gate
```

```python
def test_logic(logic_gate):
    for x1, x2 in [(0, 0), (0, 1), (1, 0), (1, 1)]:
        y = logic_gate(x1, x2)
        print(x1, x2, '|', y)
```

```python
AND = make_logic_gate(w=np.array([0.5, 0.5]), b=0.7)
```

```python
test_logic(AND)
```

```python
OR = make_logic_gate(w=np.array([0.5, 0.5]), b=0.2)
```

```python
test_logic(OR)
```

```python
NAND = make_logic_gate(w=np.array([-0.5, -0.5]), b=-0.7)
```

```python
test_logic(NAND)
```

```python
def net_input(X, w, b):
    return np.dot(X, w) + b
```

```python
def activation(z):
    return np.where(z > 0, 1, -1)
```

```python
z = net_input(x, w, b)
```

```python
z
```

```python
y = activation(z)
```

```python
y
```

```python
def fit(X, y, 훈련횟수=10, 학습률=0.1):
    w = np.array([0.] * X.shape[1])
    b = 0.0
    
    error_history = []
    for i in range(훈련횟수):
        print('훈련횟수: {}'.format(i+1))
        sum_square_error = 0
        for xi, yi in zip(X, y):
            zi = net_input(xi, w, b)
            yi_pred = activation(zi)
            # 가중치 갱신
            error = yi - yi_pred
            sum_square_error += error **2
            update = 학습률 * error
            w += update * xi
            b += update
        print('w: {}\tb: {}\tError: {}'.format(
            w, b, sum_square_error))
        error_history.append(sum_square_error)
        
    return w, b, error_history
```
